```python
# копия данных
data2 = data
```

```python
X_std = data2['xaxlb'].std()
X_mean = data2['xaxlb'].mean()

for i in data2['xaxlb']:
    temp = int((i - X_mean)/X_std)
    if abs(temp)>2:
#         тут должен быть код, который найденный выброс 
#         заменит на среднее значение
        data2['xaxlb'][i] = X_mean
```

```python
F_std = data2['fxqqe'].std()
F_mean = data2['fxqqe'].mean()

for i in data2['fxqqe']:
    temp = int((i - F_mean)/F_std)
    if abs(temp)>0:
#         тут должен быть код, который найденный выброс 
#         заменит на среднее значение
        data2['fxqqe'] = F_mean
```

```python
L_std = data2['lusyw'].std()
L_mean = data2['lusyw'].mean()

for i in data2['lusyw']:
    temp = int((i - L_mean)/L_std)
    if abs(temp)>0:
#         тут должен быть код, который найденный выброс 
#         заменит на среднее значение
        data2['lusyw'] = L_mean
        print(i)
```
